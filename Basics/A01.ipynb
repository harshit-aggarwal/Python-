```python
# Task 4

# Import random to use to pick randomly flip outcome
import random

# A funtion which simulates a coin toss given the number of coin-toss as an input
# Function prints the number of heads, tails and the probability of each
def toss(number):
    
    # Empty list to store outcome 
    head = []
    tails = []
    
    # Only 2 possibilities Heads ("H") or Tails ("T")
    choices = ["H","T"]
    
    # Run a loop to simulate coin-toss according to user input 
    for i in range(number):
        choice = random.choice(choices)
        # If head add to heads list
        if choice == "H":
            head.append(i)
        else:
        # Otherwise add to tails list
            tails.append(i)
    print("Simulating", number, "coin toss")
    # print the outcomes
    print("Number of heads:", len(head))
    print("Number of tails:" ,len(tails))
    print("Probability of heads:", len(head)/number)
    print("Probability of tails:", len(tails)/number)

    
# Simulate for 100 coin toss
toss(100)

# Simulate for 500 coin toss
toss(500)

```

```python
import random

# A function to simulate 2 dice rolls and return the sum of the rolls
def getSum():
    a = random.randint(1,6)
    b = random.randint(1,6)
    tot = a + b
    
    print("Your roll is:", tot)
    return(tot)

# A function to handle the game if it enters the point game situation
# Input is the "point" of the player
def point_game(n):
    # Roll again
    roll2 = getSum()
    play2 = 1
    # Play point game
    while play2 == 1:
        # If second roll is 7, the player automatically loses, end game
        if roll2 == 7:
            print("You lose!")
            break
        # If subsuquent roll is the point, the player wins, end game
        elif roll2 == n:
            print("You made your point, you win!")
            break
        # If neither 7 nor point is reached, roll again
        else:
            roll2 = getSum()

            
play = 1

# Play the game
while play == 1:
    # Get the first sum 
    roll = getSum()
    # If the sum is 7 or 11, the player automatically wins, end game
    if roll == 7 or roll == 11:
        print("You win!")
        play = 0
    # If the sum is 2,3 or 12, the player automatically loses, end game
    elif roll == 2 or roll == 3 or roll == 12:
        print("My game, You lose!")
        play = 0
    # If neither situation then we enter point game part
    else:
        print("Point Game, Your point is:", roll)
        print("Roll again")
        point_game(roll)
        play = 0
```
